```python
#SAMPLE SET
albumSet1= set(['thriller','ac/dc','black in black'])
albumSet2= set (['ac/dc','black in black','the dark side of the moon'])
```

```python
albumSet1,albumSet2
```

```python
#finding intersection
intersection= albumSet1 & albumSet2
intersection
```

```python
#finding differncce in set 1 but not in set 2
albumSet1.difference(albumSet2)
```

```python
albumSet2.difference(albumSet1)
```

```python
# Use intersection method to find the intersection of album_list1 and album_list2

albumSet1.intersection(albumSet2)   
```

```python
#finding the union of twoo sets
albumSet1.union(albumSet2)
```

```python
#checking weather superset or not
set(albumSet1).issuperset(albumSet2)
```

```python
#CHECKING FOR SUBSET
set(albumSet2).issubset(albumSet1)
```

```python
#again checking
set({'black in black','ac/dc'}).issubset(albumSet1)
```
